```python
pcos=pcos.dropna()
pcos.isnull().sum()
```

```python
x = pcos.drop('PCOS (Y/N)',axis=1)
y = pcos['PCOS (Y/N)']
```

```python
pcos.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
def models(x_train,y_train):
    #Logistic Regression
    from sklearn.linear_model import LogisticRegression
    model = LogisticRegression(random_state=0)
    model.fit(x_train,y_train)

    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    classifier = RandomForestClassifier(n_estimators=50,criterion="entropy",random_state=0)
    classifier.fit(x_train,y_train)
    
    print("Logistic Regression Accuracy",model.score(x_train,y_train))
    print("Random Forest Classifier Accuracy",classifier.score(x_train,y_train))

    return model,classifier
```

```python
m = models(x_train,y_train)
m
```

```python
from sklearn.metrics import confusion_matrix
for i in range(len(m)):
    cm = confusion_matrix(y_test,m[i].predict(x_test))
    tp = cm[1][1]
    tn = cm[0][0]
    fn = cm[1][0]
    fp = cm[0][1]
    print(cm)
    print("testing accuracy = ",(tp+tn)/(tp+tn+fp+fn))
```
